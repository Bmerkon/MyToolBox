### 处理一下分钟数据

```python
import numpy as np
datamin['month']=datamin['ticker'].str.slice(start=4)
datamin['year_end']=datamin['ticker'].str.slice(start=3,stop=4)
datamin=datamin[datamin['year_end'].str.isnumeric()]
datamin['rank']=datamin['month'].astype(int).values+np.mod(datamin['year_end'].astype(int).values,2)*12
```

```python
datamin.head(5)
```

```python
dp_open=datamin.pivot_table(index='datetime',columns='rank',values='open').tz_convert('Asia/Hong_Kong',axis=0)
dp_oi=datamin.pivot_table(index='datetime',columns='rank',values='openinterest').tz_convert('Asia/Hong_Kong',axis=0)
```
