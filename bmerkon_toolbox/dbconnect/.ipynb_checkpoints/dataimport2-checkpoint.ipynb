```python
def  data2mat3(innercode_list,tradingdate,innercode,startdate,enddate,datadate,data):
#    pdb.set_trace()
    ndata=len(innercode)
    ndate=len(tradingdate)
    nstock=len(innercode_list)
    ind_notdrop=np.array([True]*ndata)
    innercode_list=innercode_list.astype(int)
    code_list=innercode.unique()
    for code in code_list:
        indcode=innercode==code
        cummaxdate=pd.Series.cummax(datadate[indcode])
        ind_notdrop[indcode]=datadate[indcode]>=cummaxdate
        
    innercode_array=innercode.values[ind_notdrop]
    startdate_array=startdate.values[ind_notdrop]
    enddate_array=enddate.values[ind_notdrop]
    data_array=data.values[ind_notdrop]
    tradingdate_array=tradingdate.values
    ndata=len(innercode_array)
    
    result_tab=np.ones([ndate,nstock])*np.nan
    #tradingdate = pd.Series(list(map(pd.Timestamp, tradingdate) ))
    for index in range(ndata):
        #print ("row[startdate] = ", row['startdate'] , " type is ", type(row['startdate']))
        #print ("tradingdate = ", tradingdate, ", tpe is", type(tradingdate[0]))        
        ind_date=np.nonzero((tradingdate_array>=startdate_array[index]) &( tradingdate_array<=enddate_array[index]))[0]
        ind_stock=np.nonzero(innercode_list==innercode_array[index])[0]
        #pdb.set_trace()
        if (len(ind_date)>0) & (len(ind_stock)>0):
            result_tab[ind_date,ind_stock]=data_array[index]
            
    result_tab=pd.DataFrame(index=tradingdate,columns=innercode_list,data=result_tab)        
    return result_tab
```

```python
import pdb
import numpy as np
def data2mat(innercode_list,tradingdate,innercode,startdate,datadate,data):

    # 处理有效数据
    code_list=innercode.unique()
    ndata=len(innercode)
    ind_notdrop=np.array([True]*ndata)
    for code in code_list:
        indcode=innercode==code
        cummaxdate=pd.Series.cummax(datadate[indcode])
        ind_notdrop[indcode]=datadate[indcode]>=cummaxdate
        
    temp_tab=pd.DataFrame()
    temp_tab['innercode']=innercode[ind_notdrop]
    temp_tab['startdate']=startdate[ind_notdrop]
    temp_tab['data']=data[ind_notdrop]
    
    date_list=temp_tab['startdate'].unique()
    
    # 增加空白数据
        
    #pdb.set_trace()
    alldate=list(set(np.hstack((tradingdate,date_list)))).sort()
    extracode=set(innercode_list)-set(code_list)
    k=list(extracode)
    
    result=temp_tab.pivot_table(index='startdate',columns='innercode',values='data',aggfunc=np.sum)
    pdb.set_trace()
    result[list(extracode)]=np.nan
    result=result.reindex(alldate)

   # result=temp_tab.pivot(index='startdate',columns='innercode',values='data')
    return result
    
        
```

```python
import datetime
import pdb
def data2mat2(innercode_list,tradingdate,innercode,startdate,enddate,datadate,data):
#    pdb.set_trace()
    ndata=len(innercode)
    ndate=len(tradingdate)
    nstock=len(innercode_list)
    ind_notdrop=np.array([True]*ndata)
    innercode_list=innercode_list.astype(int)
    code_list=innercode.unique()
    for code in code_list:
        indcode=innercode==code
        cummaxdate=pd.Series.cummax(datadate[indcode])
        ind_notdrop[indcode]=datadate[indcode]>=cummaxdate
        
    temp_tab=pd.DataFrame()
    temp_tab['innercode']=innercode[ind_notdrop]
    temp_tab['startdate']=startdate[ind_notdrop]
    temp_tab['enddate']=enddate[ind_notdrop]
    temp_tab['data']=data[ind_notdrop]
    
    
    result_tab=np.ones([ndate,nstock])*np.nan
    #tradingdate = pd.Series(list(map(pd.Timestamp, tradingdate) ))
    for index,row in temp_tab.iterrows():
        #print ("row[startdate] = ", row['startdate'] , " type is ", type(row['startdate']))
        #print ("tradingdate = ", tradingdate, ", tpe is", type(tradingdate[0]))        
        ind_date=np.nonzero((tradingdate>=row['startdate']) &( tradingdate<=row['enddate']))[0]
        ind_stock=np.nonzero(innercode_list==row['innercode'])[0]
        #pdb.set_trace()
        if (len(ind_date)>0) & (len(ind_stock)>0):
            result_tab[ind_date,ind_stock]=row['data']
            
    result_tab=pd.DataFrame(index=tradingdate,columns=innercode_list,data=result_tab)        
    return result_tab
            
```
